```python
ClickedCounter = 0
UnClickedCounter = 0
AgeOfClicked = 0
AgeOfUnClicked = 0
meanOfClicked = 0
meanOfUnClicked = 0
tic = time.time()
for i in range(len(advertising_df)): 
    if (advertising_df.loc[i, "Clicked on Ad"] == 1):
        AgeOfClicked += advertising_df.loc[i, "Age"]
        ClickedCounter +=1 
    if (advertising_df.loc[i, "Clicked on Ad"] == 0):
        AgeOfUnClicked += advertising_df.loc[i, "Age"]
        UnClickedCounter += 1
meanOfClicked = AgeOfClicked/ClickedCounter
meanOfUnClicked = AgeOfUnClicked/UnClickedCounter
toc = time.time()
print("meanOfClicked: %f " % meanOfClicked)
print("meanOfUnClicked: %f " % meanOfUnClicked)
print("for loop: %f ms" % ((toc - tic) * 1000))

```

```python
df_S9 = advertising_df.drop(['Ad Topic Line', 'City' , 'Gender' , 'Country' , 'Timestamp' , 'Unnamed: 0'], axis=1)
```

```python
df_S9.head(15)
```

```python
def z_score(df):
    df_std = df.copy()
    for column in df_std[['Daily Time Spent on Site', 'Age' , 'Area Income', 'Daily Internet Usage']]:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std
    
df_ad_standardized = z_score(df_S9)

df_ad_standardized
```
